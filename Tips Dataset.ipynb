```python
sns.displot(df1['total_bill'])
```

```python
a=df1['total_bill']
mean=a.mean()
median=np.median(a)
mode=a.mode()
```

```python
sns.distplot(a,hist=False)
plt.axvline(mean,color='k',label='mean')
plt.axvline(median,color='r',label='median')
plt.axvline(mode[0],color='g',label='mode')
plt.legend()
plt.show();
```

```python
a.plot(kind='kde')
plt.axvline(mean,color='g',label='mean')
plt.axvline(median,color='y',label='median')
plt.axvline(mode[0],color='b',label='mode')
plt.legend
```

```python
sns.stripplot(x='day',y='total_bill',data=df1,jitter=False)
plt.grid()
plt.axhline(20,color='black')
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=df1,hue='sex')
plt.grid()
plt.axhline(20,color='black')
plt.show()
```
